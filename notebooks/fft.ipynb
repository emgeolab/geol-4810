```python
import numpy as np
import matplotlib.pyplot as plt

# -----------------------------
# Sampling parameters
# -----------------------------
fs = 100.0          # sampling frequency (Hz)
s = 1 / fs          # sampling interval
fn = 1./ (2*s)
N = 1000             # number of samples
t = np.arange(N) * s

# -----------------------------
# Signal (time domain)
# -----------------------------
x = (
    2.0 * np.cos(2 * np.pi * 5 * t)   # 5 Hz cosine
    + 1.0 * np.sin(2 * np.pi * 12 * t) # 12 Hz sine
)

# -----------------------------
# FFT
# -----------------------------
X = np.fft.fft(x)
freq = np.fft.fftfreq(N, d=s)

# -----------------------------
# Plot time domain
# -----------------------------
plt.figure(figsize=(10, 4))
plt.plot(t, x)
plt.xlabel("Time (s)")
plt.ylabel("Amplitude")
plt.title("Time-domain signal")
plt.grid(True)
plt.tight_layout()
plt.show()

# -----------------------------
# Plot magnitude spectrum
# -----------------------------
plt.figure(figsize=(10, 4))
plt.stem(freq, np.abs(X), basefmt=" ")
plt.xlim(0, 20)   # show positive frequencies only
plt.xlabel("Frequency (Hz)")
plt.ylabel("|X(f)|")
plt.title("FFT magnitude spectrum")
plt.grid(True)
plt.tight_layout()
plt.show()
```
